##시장 상황에 따른 소비자 유형 힘의 변화분석


1. 문제제기자 (Problem Recognizer) 문제제기자는 소비자들 사이에서 특정 요구나 문제점을 인식하는 중요한 역할을 합니다. 현재 시장에서 건강식에 대한 관심이 크게 증가하는 추세이며, 많은 소비자들이 체중 관리와 건강한 식습관을 추구하고 있습니다. 이로 인해 전통적인 김치에서 저염김치, 유기농 재료로 만든 김치와 같은 건강 지향적인 옵션에 대한 필요성이 더욱 부각되고 있습니다.
특히, 최근의 연구나 미디어에서 “김치가 건강에 미치는 긍정적 효과”에 대한 정보가 확산됨에 따라, 건강한 김치 소비를 위한 요구가 더욱 증가하고 있습니다.

2. 영향자 (Influencer) 영향자는 소비자들이 제품 선택 시 큰 영향을 미치는 인물들입니다. 최근 들어 소셜 미디어의 영향력이 커짐에 따라, 건강식품에 대한 권위 있는 사용자의 추천이나 인플루언서의 리뷰가 더욱 중요해졌습니다. 대원 종가집 김치의 경우, 유명한 셰프나 건강 관련 블로거가 이 제품을 추천하면, 소비자들의 구매 의사에 직접적인 영향을 미치게 됩니다. 소비자들은 이러한 영향자들의 의견을 신뢰하며, 그들의 추천에 따라 행동하는 경향이 강해지고 있습니다.

3. 결정자 (Decider) 결정자는 최종적으로 구매 결정을 내리는 소비자를 의미합니다. 대원 종가집 김치의 경우, 주로 가정에서 가족의 식사를 준비하는 사람들, 특히 가정주부가 결정자가 됩니다. 이들은 제품의 가격, 품질, 영양 성분 등을 중요한 기준으로 삼습니다. 최근에는 가격 대비 품질을 중시하고, 브랜드의 신뢰성도 함께 고려하게 되는데, 브랜드가 소비자에게 신뢰를 줄 수 있는 요소가 필요합니다. 예를 들어, 대원 종가집은 전통 방식으로 제조된 품질 높은 김치를 제공하면서 인증 받은 원재료를 사용하고 있다는 점을 강조할 수 있습니다.

4. 구매자 (Buyer) 구매자는 실제로 제품을 구매하는 소비자들입니다. 대원 종가집 김치의 구매자들은 연령대와 생활 스타일에 따라 다양화되고 있습니다. 건강을 중시하는 젊은 소비자들, 전통적인 한국 음식을 선호하는 중장년층, 그리고 자녀를 둔 가정주부까지, 각기 다른 소비자들이 있습니다. 이들은 맛, 가격, 그리고 건강에 대한 정보에 기반하여 제품을 선택샙니다. 따라서 대원 종가집은 다양한 마케팅 전략을 통해 각 소비자 그룹의 니즈에 잘 대응해야 합니다.

5. 사용자 영향자 (User Influencer) 사용자 영향자는 제품을 실제로 사용하고 경험한 후, 그 경험을 다른 소비자와 공유하는 사람들을 의미합니다. 대원 종가집 김치의 경우, 소비자들은 자신이 사용한 후의 경험을 소셜 미디어 플랫폼이나 개인 블로그를 통해 공유할 수 있으며, 이는 다른 소비자들에게 강력한 구매 동기를 제공합니다. 사용자의 리뷰와 평가는 신규 소비자에게 제품 구매의 결정적인 이유가 되기도 합니다. 특히, 긍정적인 경험을 공유한 소비자들은 브랜드 충성도를 심화시킬 수 있습니다. 결론 PSPS 분석을 통해 대원 종가집 김치는 시장에서 다양한 소비자 유형의 요구와 행동 양식에 대해 깊이 이해하고 이를 반영하는 전략을 수립할 수 있습니다. 각 요소(문제제기자, 영향자, 결정자, 구매자, 사용자 영향자)가 상호작용하면서 소비자의 구매 결정에 영향을 주기 때문에, 대원 종가집은 이러한 흐름을 파악하고 그에 맞는 맞춤형 마케팅과 제품 개발 전략을 구사해야 합니다. 결국, 대원 종가집 김치는 전통적인 가치와 현대적인 소비 트렌드를 조화롭게 결합하여 소비자의 건강과 맛, 그리고 신뢰를 동시에 충족시키는 제품으로 자리매김할 수 있을 것입니다.

##마케팅 전략 솔루션

1. 건강 지향적 제품 라인 확대:

저염 김치, 유기농 재료를 사용한 김치 등 건강에 초점을 맞춘 제품군을 개발하고 확대합니다.
김치의 건강 효능에 대한 연구 결과를 적극적으로 홍보합니다.


2. 인플루언서 마케팅 강화:

유명 셰프, 건강 전문가, 식품 블로거 등과 협력하여 제품을 홍보합니다.
소셜 미디어를 통한 인플루언서 캠페인을 진행합니다.


3. 타겟 소비자별 맞춤 마케팅:

가정주부, 젊은 세대, 중장년층 등 다양한 소비자 그룹에 맞는 마케팅 메시지를 개발합니다.
각 그룹의 니즈에 맞는 제품을 홍보합니다 (예: 젊은 층을 위한 간편 포장, 중장년층을 위한 전통 레시피 강조).


4. 브랜드 신뢰성 강화:

전통 방식의 제조 과정, 품질 관리 시스템, 원재료의 안전성 등을 강조합니다.
투명한 정보 공개로 소비자 신뢰를 구축합니다.


5. 사용자 경험 공유 플랫폼 구축:

공식 웹사이트나 앱에서 소비자들이 제품 리뷰를 쉽게 남길 수 있는 플랫폼을 만듭니다.
긍정적인 사용자 경험을 소셜 미디어에서 공유하도록 장려합니다.


6. 교육 콘텐츠 제작:

김치의 영양학적 가치, 건강 효능, 다양한 활용법 등에 대한 교육 콘텐츠를 제작하여 배포합니다.


7. 맞춤형 구독 서비스:

소비자의 선호도에 따라 다양한 김치를 정기적으로 배송하는 구독 서비스를 제공합니다.

## PSPS(Power Shift Pentagon Structure Analysis)

- 문제제기자 (Problem Recognizer): 8점

    - 건강식에 대한 관심 증가와 저염, 유기농 김치 수요 인식
    - 김치의 건강 효과에 대한 정보 확산 인지


- 영향자 (Influencer): 7점

    - 소셜 미디어 인플루언서의 영향력 인식
    - 셰프, 건강 블로거 등의 추천 중요성 파악


- 결정자 (Decider): 9점

    - 주 고객층인 가정주부의 구매 결정 과정 이해
    - 가격, 품질, 영양 성분, 브랜드 신뢰성 등 고려 요소 파악


- 구매자 (Buyer): 8점

    - 다양한 연령대와 생활 스타일의 구매자 특성 인식
    - 맛, 가격, 건강 정보에 기반한 선택 기준 이해


- 사용자 영향자 (User Influencer): 7점

    - 소셜 미디어, 블로그 등을 통한 사용 경험 공유 중요성 인식
    - 사용자 리뷰가 신규 소비자 유치에 미치는 영향 파악

## 추가 분석

####KOSIS 통계청 데이터를 사용하여 식품소비 트렌드별 동조성 조사에 관한 결과를 Radar Graph 로 분석하였습니다.

1. 첫번째 자료 : 2018년 김치류와 절임류에 대한 소비자 성향 조사
2. 두번째 자료 : 2019년과 2020년의 김치류와 절임류에 대한 소비자 성향 조사
3. 세번째 자료 : 2021년, 2022년, 2023년의 김치류와 절임류에 대한 소비자 조사

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/식품소비_트렌드별_동조성_김치류__절임류__20241002034817.xlsx'
df = pd.read_excel(file_path)


df_filtered = df.iloc[3:, :]
df_filtered.columns = df.iloc[1, :]
df_filtered = df_filtered.reset_index(drop=True)

df_kimchi = df_filtered.iloc[0::2]
df_pickled = df_filtered.iloc[1::2]

kimchi_values = df_kimchi.iloc[0, 2:].astype(float).values
pickled_values = df_pickled.iloc[0, 2:].astype(float).values
categories = df_filtered.columns[2:]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  #

kimchi_values = np.append(kimchi_values, kimchi_values[0])
pickled_values = np.append(pickled_values, pickled_values[0])

ax.plot(angles, kimchi_values, linewidth=2, linestyle='solid', label='김치류', color='blue')
ax.fill(angles, kimchi_values, color='blue', alpha=0.25)

ax.plot(angles, pickled_values, linewidth=2, linestyle='solid', label='절임류', color='green')
ax.fill(angles, pickled_values, color='green', alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=10)

plt.title('2018년 김치류 및 절임류 소비자 성향 비교', size=15, y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
```

위의 레이더 차트는 2018년 김치류와 절임류에 대한 소비자 성향을 비교한 결과입니다.

###분석 내용
1. 소비자 성향의 전반적 특징:

  - 두 제품군 모두 "가격이 비싸도 건강에 좋은 원료, 안전성이 확보된 제품을 구입하겠다"라는 항목에 대해 높은 선호도를 보이고 있습니다. 이는 소비자들이 건강과 안전을 우선시하는 경향이 강하다는 것을 나타냅니다.
  - 반면 "가격이 비싸도 프리미엄(고급) 제품을 구입하겠다" 항목에서는 상대적으로 낮은 점수를 보였습니다. 이는 소비자들이 프리미엄 제품에 대한 선호도는 다소 낮지만, 건강과 안전에 대한 가치가 높은 경우에는 더 많은 비용을 지불할 의향이 있다는 것을 의미합니다.

2. 김치류와 절임류의 비교:

  - 김치류는 절임류에 비해 "가격이 비싸도 다양한 맛을 가진 제품을 구입하겠다" 항목에서 상대적으로 높은 선호도를 보입니다. 이는 소비자들이 김치류에서 더 다양한 맛을 기대하고 있다는 것을 의미합니다.
  - 반면 절임류는 "제품의 업그레이드/신제품 개발과 관계없이 가격만 저렴하면 구입하겠다" 항목에서 더 높은 선호도를 보이고 있습니다. 이는 절임류 소비자들이 가격에 더 민감하게 반응하며, 새로운 제품보다는 기존 제품에 익숙해 있다는 것을 나타냅니다.

3. 소비 패턴의 차이점:

  - 김치류는 건강 및 맛의 다양성에 대한 요구가 더 강한 반면, 절임류는 가격과 편리성에 대한 요구가 강합니다.
  - 이는 김치류가 전통적인 식품이면서도 건강을 강조하는 이미지가 강하기 때문에 소비자들이 더욱 건강과 맛에 집중하는 것으로 해석할 수 있습니다.

###결론
  - 김치류: 건강, 원료의 안전성, 맛의 다양성에 대한 소비자 요구가 높으므로, 고품질의 원료와 다양한 맛을 강조하는 마케팅 전략이 효과적일 것입니다.
  - 절임류: 가격 경쟁력을 강조하고, 기존 제품의 편리성과 익숙함을 유지하면서 새로운 맛이나 저렴한 가격에 대한 프로모션을 진행하는 것이 효과적일 수 있습니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

new_file_path = '/content/식품소비_트렌드별_동조성_김치류__절임류__20241002034757.xlsx'
df_new = pd.read_excel(new_file_path)

df_filtered_new = df_new.iloc[2:, :]
df_filtered_new.columns = df_new.iloc[1, :]
df_filtered_new = df_filtered_new.reset_index(drop=True)

df_kimchi_new = df_filtered_new.iloc[0::2]
df_pickled_new = df_filtered_new.iloc[1::2]

kimchi_values_2019 = df_kimchi_new.iloc[0, 2:8].astype(float).values
pickled_values_2019 = df_pickled_new.iloc[0, 2:8].astype(float).values
categories_2019 = df_filtered_new.columns[2:8]

kimchi_values_2020 = df_kimchi_new.iloc[0, 10:16].astype(float).values
pickled_values_2020 = df_pickled_new.iloc[0, 10:16].astype(float).values
categories_2020 = df_filtered_new.columns[10:16]

fig, ax = plt.subplots(1, 2, figsize=(18, 8), subplot_kw=dict(polar=True))

angles_2019 = np.linspace(0, 2 * np.pi, len(categories_2019), endpoint=False).tolist()
angles_2019 += angles_2019[:1]

angles_2020 = np.linspace(0, 2 * np.pi, len(categories_2020), endpoint=False).tolist()
angles_2020 += angles_2020[:1]

kimchi_values_2019 = np.append(kimchi_values_2019, kimchi_values_2019[0])
pickled_values_2019 = np.append(pickled_values_2019, pickled_values_2019[0])

kimchi_values_2020 = np.append(kimchi_values_2020, kimchi_values_2020[0])
pickled_values_2020 = np.append(pickled_values_2020, pickled_values_2020[0])

ax[0].plot(angles_2019, kimchi_values_2019, linewidth=2, linestyle='solid', label='김치류', color='blue')
ax[0].fill(angles_2019, kimchi_values_2019, color='blue', alpha=0.25)

ax[0].plot(angles_2019, pickled_values_2019, linewidth=2, linestyle='solid', label='절임류', color='green')
ax[0].fill(angles_2019, pickled_values_2019, color='green', alpha=0.25)

ax[0].set_xticks(angles_2019[:-1])
ax[0].set_xticklabels(categories_2019, fontsize=10)
ax[0].set_title('2019년 김치류 및 절임류 소비자 성향 비교', size=14, y=1.1)
ax[0].legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

ax[1].plot(angles_2020, kimchi_values_2020, linewidth=2, linestyle='solid', label='김치류', color='blue')
ax[1].fill(angles_2020, kimchi_values_2020, color='blue', alpha=0.25)

ax[1].plot(angles_2020, pickled_values_2020, linewidth=2, linestyle='solid', label='절임류', color='green')
ax[1].fill(angles_2020, pickled_values_2020, color='green', alpha=0.25)

ax[1].set_xticks(angles_2020[:-1])
ax[1].set_xticklabels(categories_2020, fontsize=10)
ax[1].set_title('2020년 김치류 및 절임류 소비자 성향 비교', size=14, y=1.1)
ax[1].legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

plt.tight_layout()
plt.show()
```

#### 절임류: 가격 민감도가 더욱 높아졌으므로, 저렴한 가격을 유지하면서도 품질에 대한 신뢰성을 전달하는 것이 중요합니다.

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt

datalab_file_path = '/content/datalab.xlsx'
df_datalab = pd.read_excel(datalab_file_path)

df_datalab.columns = ['날짜', '김치', '날짜_종가집', '종가집', '날짜_포장김치', '포장김치', '날짜_포기김치', '포기김치']

df_datalab_cleaned = df_datalab[['날짜', '김치', '종가집', '포장김치', '포기김치']]

df_datalab_cleaned['날짜'] = pd.to_datetime(df_datalab_cleaned['날짜'])
df_datalab_cleaned['김치'] = pd.to_numeric(df_datalab_cleaned['김치'], errors='coerce')
df_datalab_cleaned['종가집'] = pd.to_numeric(df_datalab_cleaned['종가집'], errors='coerce')
df_datalab_cleaned['포장김치'] = pd.to_numeric(df_datalab_cleaned['포장김치'], errors='coerce')
df_datalab_cleaned['포기김치'] = pd.to_numeric(df_datalab_cleaned['포기김치'], errors='coerce')

df_datalab_cleaned.set_index('날짜', inplace=True)

plt.figure(figsize=(14, 7))
plt.plot(df_datalab_cleaned.index, df_datalab_cleaned['김치'], label='김치', color='blue')
plt.plot(df_datalab_cleaned.index, df_datalab_cleaned['종가집'], label='종가집', color='green')
plt.plot(df_datalab_cleaned.index, df_datalab_cleaned['포장김치'], label='포장김치', color='orange')
plt.plot(df_datalab_cleaned.index, df_datalab_cleaned['포기김치'], label='포기김치', color='red')

plt.title('2020년 10월 ~ 2024년 10월 김치 관련 검색량 추세')
plt.xlabel('날짜')
plt.ylabel('검색량')
plt.legend(loc='upper left')
plt.tight_layout()

plt.show()
```
